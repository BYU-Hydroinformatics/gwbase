# Load Streamflow ML results

## Load streamflow data (bfd=1 periods)

```python
import os
from gwbase.data_loading import load_streamflow_data

streamflow = load_streamflow_data('../data/raw/streamflow/GSLB_ML')

os.makedirs('../data/processed/streamflow', exist_ok=True)
streamflow.to_csv('../data/processed/streamflow/q_bfd_1.csv', index=False)
streamflow.head()
```

# PCHIP well wte

```python
import pandas as pd
well_ts = pd.read_csv('../data/raw/groundwater/GSLB_1900-2023_TS_with_aquifers.csv')
well_ts.columns = well_ts.columns.str.lower()
well_ts.info()
```

## Clean groundwater data for interpolation

```python
from gwbase.preprocessing import clean_well_data_for_interpolation

clean_data = clean_well_data_for_interpolation(well_ts)
```

## PCHIP wte

```python
import pandas as pd
from gwbase.interpolation import interpolate_with_well_info

well_info = pd.read_csv('../data/raw/groundwater/GSLB_1900-2023_wells_with_aquifers.csv')

merged_data = interpolate_with_well_info(clean_data, well_info)
merged_data.to_csv('../data/processed/well_pchip.csv', index=False)
merged_data.head()
```
